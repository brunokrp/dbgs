```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn
import seaborn as sns
import numpy as np
%matplotlib inline


#functions
def gravGroup(x):
   if x ==1 or x ==4:
       return 1
   elif x ==2 or x ==3:
       return 0

def locPedest(x):
   if x == 0:
       return 0
   else:
       return 1
    
def ageCateg18(x):
   if x < 18:
       return 1
   else:
        return 0
    
def ageCateg65(x):
   if x > 65: 
       return 1
   else:
        return 0
    
users = pd.read_csv('../dataset/users.csv', low_memory=False, sep=',', encoding='utf-8')
users = users.drop(columns=['actp', 'etatp', 'place', 'trajet', 'secu'])

    
users['grav'] = users['grav'].apply(gravGroup)
users['locp'] = users['locp'].apply(locPedest)
carac = pd.read_csv("../dataset/caracteristics.csv", encoding="latin-1", low_memory=False).sort_values('Num_Acc')

usr_carac = pd.merge(users, carac, how='inner', on='Num_Acc', left_on=None, right_on=None, left_index=False, right_index=False, sort=True, suffixes=('_us', '_cr'), copy=True, indicator=False,validate=None)
users[users['an_nais'].isna()]['an_nais'] = users['an_nais'].mean()
usr_carac['age'] = (2000 + usr_carac['an']) - usr_carac['an_nais']
usr_carac['ageCateg18'] = usr_carac['age'].apply(ageCateg18)
usr_carac['ageCateg65'] = usr_carac['age'].apply(ageCateg65)

```
